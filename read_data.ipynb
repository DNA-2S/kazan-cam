```python
import glob
import cv2
from matplotlib import pyplot as plt
import matplotlib.patches as patches


dirs = [i for i in range(2, 16)]
dataset = []
for dr in dirs:
    for file in glob.glob(f'dataset/{dr}/*.txt'):
        with open(file, 'r') as f:
            data = f.read()
            data = data.split('\n')
            data = [d.split() for d in data]
            data2 = []
            for d in data:
                lst = []
                for i in d:
                    lst.append(float(i))
                data2.append(lst)
            
        dataset.append([dr, file.replace(".txt",".png"), data2])
        break
    break


print(dataset)
# plt.figure(figsize=(40, 8))
# plt.imshow(img)
# plt.show()

img = cv2.imread(dataset[0][1])


# Create figure and axes
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5)
# Display the image
ax.imshow(img)


h,w,c = img.shape
coords = [0, 0.576953125, 0.39975925925925926, 0.039000000000000055, 0.0676111111111111]

if coords[0] == 0:
    color = 'g'
else:
    color = 'r'
x1 = coords[1] * w
y1 = coords[2] * h
w = coords[3] * w
h = coords[4] * h

x1 = x1 - 0.5*w
y1 = y1 - 0.5*h

# Create a Rectangle patch
rect = patches.Rectangle((x1, y1), w, h, linewidth=1, edgecolor=color, facecolor='none')

# Add the patch to the Axes
ax.add_patch(rect)

plt.show()
```
